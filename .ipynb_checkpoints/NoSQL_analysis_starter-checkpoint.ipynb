### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene":20}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(count)
# Display the first document in the results using pprint
first_result = establishments.find_one(query)
pprint(first_result)
```

```python
# Convert the result to a Pandas DataFrame
#import pandas
import pandas as pd

hyg_20_query_results = establishments.find(query)
hyg_20_query_results_list = list(hyg_20_query_results)

hyg_20_df = pd.DataFrame(hyg_20_query_results_list)

# Display the number of rows in the DataFrame
row_count = hyg_20_df.shape[0]

# Display the first 10 rows of the DataFrame
hyg_20_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'AddressLine3' : 'London',
         'RatingValue' : {'$gte': 4}}
#there are no 'London' LocalAuthorities See cell below

# Use count_documents to display the number of documents in the result
row_count = establishments.count_documents(query)
print(row_count)
# Display the first document in the results using pprint
first_result = establishments.find_one(query)
first_result
```

```python
local_authorities = establishments.distinct('LocalAuthorityName')
local_authorities
```

```python
# Convert the result to a Pandas DataFrame
query_list = list(establishments.find(query))
london_df = pd.DataFrame(query_list)

# Display the number of rows in the DataFrame
row_count = london_df.shape[0]
print(row_count)
# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
from bson.objectid import ObjectId
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
penang_query = establishments.find({'_id' : ObjectId('6498d13d20193d6b1d13b1b8')})

degree_search = 0.01
latitude = penang_query[0]['geocode']['latitude']
longitude = penang_query[0]['geocode']['longitude']

query = {
    'geocode' : {
        '$geoWithin' : {
            '$center' : [[longitude, latitude], degree_search]
        }}, 'RatingValue' : 5}
sort = [('scores.Hygiene', 1)] 

# Print the results
results = establishments.find(query).sort(sort)
for doc in results:
    pprint(doc)
```

```python
# Convert result to Pandas DataFrame
results = establishments.find(query).sort(sort)

#create empty list for results to be added to
penang_list = []

#create list of results to convert to a pandas dataframe
for doc in results:
    penang_list.append(doc)
    

#create dataframe from list
nearest_to_penang_df = pd.DataFrame(penang_list)
nearest_to_penang_df
```
